#Les dictionnaires

un dictionnaire est une structure de données qui permet de stocker des paires clé-valeur. Chaque élément dans un dictionnaire est associé à une clé unique qui permet d'accéder à la valeur correspondante. Voici un exemple simple de création et d'utilisation d'un dictionnaire en Python :

````python
# Créez un dictionnaire avec quelques paires clé-valeur
mon_dictionnaire = {
    "nom": "John",
    "age": 25,
    "ville": "Paris"
}

# Accédez à la valeur en utilisant la clé
nom_utilisateur = mon_dictionnaire["nom"]
age_utilisateur = mon_dictionnaire["age"]

# Modifiez une valeur existante
mon_dictionnaire["age"] = 26

# Ajoutez une nouvelle paire clé-valeur
mon_dictionnaire["profession"] = "Ingénieur"

# Supprimez une paire clé-valeur
del mon_dictionnaire["ville"]

# Parcourez les clés et les valeurs du dictionnaire
for cle, valeur in mon_dictionnaire.items():
    print(f"{cle}: {valeur}")
````

Les méthodes sur les dictionnaires :

La méthode key() : Retourne une vue sur les clés du dictionnaire.

````python

mon_dict = {"nom":"Joyce","age":"45"}
clefs = mon_dict.keys()
print(clefs)
````

la méthode values(): Retourne une vue sur les valeurs du dictionnaire.

````python
val = mon_dict.values()
print(val)
````

La méthode items(): Retourne une vue sur les paires clé-valeur du dictionnaire.

````python
Objet = mon_dict.items()
print(Objet)
````

La méthode get(cle, valeur_par_defaut): Retourne la valeur associée à la clé spécifiée, ou une valeur par défaut si la clé n'est pas présente.

````python
age = mon_dict['age']
age
````

La méthode pop(cle, valeur_par_defaut): Supprime et retourne la valeur associée à la clé spécifiée, ou une valeur par défaut si la clé n'est pas présente

````python
age = mon_dict.pop("age",23)
age
````

La méthode update(dictionnaire): Met à jour le dictionnaire avec les paires clé-valeur d'un autre dictionnaire.

````python
second_dict = {"etat": "célibataire "," age":"25"}
mon_dict.update (second_dict)
print(mon_dict)
````

Notes

2. Syntaxe : La syntaxe pour créer un dictionnaire utilise des accolades {}. Les paires clé-valeur sont séparées par des virgules, et la clé et la valeur sont séparées par deux-points (:).

````python
mon_tuple = (1, "abc", 3.14)
mon_tuple
````

#les sets

En Python, un ensemble (set) est une structure de données qui représente une collection non ordonnée d'éléments uniques. Les sets sont utiles pour effectuer des opérations mathématiques telles que l'union, l'intersection, la différence, etc. Voici quelques caractéristiques importantes des sets :

1. Syntaxe : Les sets sont définis en utilisant des accolades {}.

````python
mon_set = {1, 2, 3, 4, 5}
mon_set.add(5)
print(mon_set)
````

2. Éléments uniques : Les sets ne peuvent pas contenir d'éléments en double. Si vous ajoutez un élément déjà présent, il ne sera pas ajouté une deuxième fois.

3. Immuabilité : Les sets sont mutables, c'est-à-dire que vous pouvez ajouter et supprimer des éléments après leur création. Cependant, les éléments individuels doivent être immuables (comme les nombres, les chaînes ou les tuples).

````python
from abc import ABC, abstractmethod

class Forme(ABC):
    @abstractmethod
    def aire(self):
        pass

class Cercle(Forme):
    def __init__(self, rayon):
        self.rayon = rayon

    def aire(self):
        return 3.14 * self.rayon**2

class Triangle(Forme):
    def __init__(self, base, hauteur):
        self.base = base
        self.hauteur = hauteur

    def aire(self):
        return 0.5 * self.base * self.hauteur

def afficher_aires(formes):
    for forme in formes:
        print(f"Aire de la forme: {forme.aire()}")

# Exemple d'utilisation
cercle = Cercle(5)
triangle = Triangle(4, 8)

afficher_aires([cercle, triangle])
````

````python
class Rectangle:
    def __init__(self, longueur, largeur):
        self.longueur = longueur
        self.largeur = largeur

    def calculer_superficie(self):
        return self.longueur * self.largeur

    def __add__(self, autre_rectangle):
        return Rectangle(self.longueur + autre_rectangle.longueur, self.largeur + autre_rectangle.largeur)

    def __eq__(self, autre_rectangle):
        return self.calculer_superficie() == autre_rectangle.calculer_superficie()
````

Générateur de Mot de Passe : Écrivez une fonction qui génère un mot de passe aléatoire de longueur donnée.

````python
import random
import string

def generer_mot_de_passe(longueur):
    caracteres = string.ascii_letters + string.digits + string.punctuation
    mot_de_passe = ''.join(random.choice(caracteres) for _ in range(longueur))
    return mot_de_passe


longueur_mot_de_passe = 10
mot_de_passe_genere = generer_mot_de_passe(longueur_mot_de_passe)
print(f"Mot de passe généré : {mot_de_passe_genere}")
````

Pierre feuille ciseaux

````python
import random

def choix_ordinateur():
    return random.choice(["pierre", "papier", "ciseaux"])

def verifier_gagnant(choix_joueur, choix_ordi):
    if choix_joueur == choix_ordi:
        return "Égalité!"
    elif (
        (choix_joueur == "pierre" and choix_ordi == "ciseaux") or
        (choix_joueur == "papier" and choix_ordi == "pierre") or
        (choix_joueur == "ciseaux" and choix_ordi == "papier")
    ):
        return "Vous avez gagné!"
    else:
        return "L'ordinateur a gagné!"

def jouer_pierre_feuille_ciseaux():
    print("Bienvenue au jeu de Pierre-Papier-Ciseaux!")
    choix_joueur = input("Choisissez pierre, papier ou ciseaux: ").lower()

    if choix_joueur not in ["pierre", "papier", "ciseaux"]:
        print("Choix invalide. Veuillez choisir pierre, papier ou ciseaux.")
        return

    choix_ordi = choix_ordinateur()

    print(f"L'ordinateur a choisi {choix_ordi}.")
    resultat = verifier_gagnant(choix_joueur, choix_ordi)
    print(resultat)

# Exécuter le jeu
jouer_pierre_feuille_ciseaux()
````

Suite de Fibonacci : Écrivez une fonction pour générer les n premiers termes de la suite de Fibonacci.

````python
def fibonacci(n):
    fib = [0, 1]
    while len(fib) < n:
        fib.append(fib[-1] + fib[-2])
    return fib
fibonacci(10)
````

Nombre Premier : Écrivez une fonction pour vérifier si un nombre est premier.

````python
def est_premier(nombre):
    if nombre < 2:
        return False
    for i in range(2, int(nombre**0.5) + 1):
        if nombre % i == 0:
            return False
    return True
est_premier(2)
````

PGCD : Écrivez une fonction pour calculer le Plus Grand Commun Diviseur (PGCD) de deux nombres

````python
def pgcd(a, b):
    while b:
        a, b = b, a % b
    return a
pgcd(20, 8)
````

Nombre Parfait : Écrivez une fonction pour vérifier si un nombre est parfait (la somme de ses diviseurs propres est égale à lui-même).

````python
def est_parfait(nombre):
    diviseurs = [i for i in range(1, nombre) if nombre % i == 0]
    return sum(diviseurs) == nombre
est_parfait (6)
````

#Les opérations Mathématiques avec NumPy

NumPy offre un large éventail d'opérations mathématiques pour travailler avec des tableaux. Voici quelques exemples d'opérations couramment utilisées :

- Addition, Soustraction, Multiplication, Division :

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

addition = a + b
soustraction = a - b
multiplication = a * b
division = a / b

print(addition)
print(soustraction)
print(multiplication)
print(division)
````

- Fonctions mathématiques universelles (ufunc) :

````python
tableau = np.array([1, 2, 3])

carre = np.square(tableau)  # Carré de chaque élément
racine_carree = np.sqrt(tableau)  # Racine carrée de chaque élément
exponentielle = np.exp(tableau)  # Exponentielle de chaque élément
print(carre)
print(racine_carree)
print(exponentielle)
````

- Produit scalaire :

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

produit_scalaire = np.dot(a, b)
produit_scalaire
````

- Somme, Moyenne, Minimum, Maximum :

````python

tableau = np.array([1, 2, 3, 4, 5])

somme = np.sum(tableau)
moyenne = np.mean(tableau)
minimum = np.min(tableau)
maximum = np.max(tableau)

print(moyenne)
print(maximum)
print(minimum)
print(somme)
````
